```python
import pandas as pd
import numpy as np
from collections import defaultdict
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Qdrant
from langchain_core.documents import Document
from qdrant_client import QdrantClient
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
import bm25s
import re

# Configuration for Qdrant
QDRANT_URL = 'https://bb3f8fb3-5dde-46d6-9961-758fa12ca505.europe-west3-0.gcp.cloud.qdrant.io'
QDRANT_API_KEY = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJhY2Nlc3MiOiJtIn0.Z1euRcj3DzrWAG7eHEiCsAxvbD-KM5vTy565y8oKqzA'
```

```python
file_path = '/kaggle/input/nlp-final/preprocessed_cv_text_data.csv'
df = pd.read_csv(file_path)
df['Combined_Text'] = df['Category'] + " " + df['Processed_Text']
df['Combined_Text'] = df['Combined_Text'].str.replace('-', ' ', regex=True)
df['Combined_Text'] = df['Combined_Text'].str.replace(r'\s+', ' ', regex=True).str.strip()
df['Combined_Text'] = df['Combined_Text'].apply(lambda text: re.sub(r'((\d+\s+){1,4}\d+|\d{8,})', '', text).strip())

def chunk_text(text, max_length=1000, overlap=200):
    chunks = []
    start = 0
    while start < len(text):
        end = start + max_length
        chunks.append(text[start:end])
        start += max_length - overlap
    return chunks

all_chunks = []
for idx, row in df.iterrows():
    chunks = chunk_text(row['Combined_Text'])
    for i, chunk in enumerate(chunks):
        all_chunks.append({
            'cv_id': idx,
            'chunk_id': i,
            'text': chunk
        })

chunk_df = pd.DataFrame(all_chunks)

# Create chunk_id_map for metadata
chunk_id_map = {i: {'cv_id': row['cv_id'], 'chunk_id': row['chunk_id']} for i, row in chunk_df.iterrows()}

# Initialize BM25 retriever
documents = chunk_df['text'].tolist()
indexed_documents = list(enumerate(documents))
corpus_only = [text for _, text in indexed_documents]
tokenized_corpus = bm25s.tokenize(corpus_only)
bm25_retriever = bm25s.BM25(corpus=corpus_only)
bm25_retriever.index(tokenized_corpus)
```

```python
# Load embedding model
embedding_function = HuggingFaceEmbeddings(model_name="BAAI/bge-m3")

# Load reranking model
tokenizer_rerank = AutoTokenizer.from_pretrained("BAAI/bge-reranker-v2-m3")
model_rerank = AutoModelForSequenceClassification.from_pretrained("BAAI/bge-reranker-v2-m3")
```

```python
# Define rerank function
def rerank(query: str, docs, top_k=3):
    query = query.strip()
    pairs = [(query, doc.page_content.replace("represent_text: ", "")) for doc in docs]
    inputs = tokenizer_rerank.batch_encode_plus(
        pairs,
        padding=True,
        truncation=True,
        return_tensors="pt",
        max_length=512,
    )
    with torch.no_grad():
        scores = model_rerank(**inputs).logits.squeeze()
    ranked = sorted(zip(docs, scores), key=lambda x: x[1], reverse=True)
    return [doc for doc, score in ranked[:top_k]]

# Define RRF fusion function
def rrf_fusion(bm25_results, embedding_results, k=10, rrf_k=60):
    scores = {}
    for rank, idx in enumerate(bm25_results):
        scores[idx] = scores.get(idx, 0) + 1 / (rrf_k + rank + 1)
    for rank, doc in enumerate(embedding_results):
        idx = int(doc.metadata["index"])
        scores[idx] = scores.get(idx, 0) + 1 / (rrf_k + rank + 1)
    ranked = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return [idx for idx, _ in ranked[:k]]

# Define search_cv_with_rrf_rerank function
def search_cv_with_rrf_rerank(query: str, top_k=10, rerank_top_k=5):
    query_emb_input = "represent_text: " + query
    client = QdrantClient(url=QDRANT_URL, api_key=QDRANT_API_KEY, prefer_grpc=False)
    qdrant_store = Qdrant(client=client, collection_name='cv-nlp', embeddings=embedding_function)
    emb_results = qdrant_store.similarity_search(query_emb_input, k=top_k)

    for doc in emb_results:
        doc.metadata["index"] = chunk_df.index[
            (chunk_df["cv_id"] == doc.metadata["cv_id"]) &
            (chunk_df["chunk_id"] == doc.metadata["chunk_id"])
        ].tolist()[0]

    query_tokens = bm25s.tokenize([query])[0]
    bm25_docs, _ = bm25_retriever.retrieve(query_tokens, k=top_k)
    bm25_top_indices = [i for i, text in indexed_documents if text in bm25_docs]

    fused_indices = rrf_fusion(bm25_top_indices, emb_results, k=top_k)
    fused_docs = []
    for idx in fused_indices:
        row = chunk_df.iloc[idx]
        text = row['text']
        meta = chunk_id_map[idx]
        meta["index"] = idx
        fused_docs.append(Document(page_content="represent_text: " + text, metadata=meta))

    reranked = rerank(query, fused_docs, top_k=rerank_top_k)

    grouped = defaultdict(list)
    for doc in reranked:
        cv_id = doc.metadata.get("cv_id")
        chunk_id = doc.metadata.get("chunk_id", -1)
        grouped[cv_id].append((chunk_id, doc.page_content.replace("represent_text: ", "")))

    cv_ids = [cv_id for cv_id in grouped.keys()]
    return cv_ids[:top_k]
```

```python
queries_df = pd.read_csv('/kaggle/input/nlp-final/generated_queries.csv')

from tqdm import tqdm

# Initialize counters for evaluation
top1_correct = 0
top5_correct = 0
top10_correct = 0
total_queries = len(queries_df)
total_queries = 500
queries_df = queries_df.iloc[:500]

# Evaluate each query
for _, row in tqdm(queries_df.iterrows(), total=total_queries, desc="Evaluating queries"):
    expected_cv_id = int(row['cv_id'])
    query = row['query']

    # Get top-10 CV IDs from RAG system
    retrieved_cv_ids = search_cv_with_rrf_rerank(query, top_k=10)

    # Check top-1
    if retrieved_cv_ids and len(retrieved_cv_ids) >= 1 and expected_cv_id == retrieved_cv_ids[0]:
        top1_correct += 1

    # Check top-5
    if len(retrieved_cv_ids) >= 5 and expected_cv_id in retrieved_cv_ids[:5]:
        top5_correct += 1
    elif len(retrieved_cv_ids) < 5 and expected_cv_id in retrieved_cv_ids:
        top5_correct += 1

    # Check top-10
    if expected_cv_id in retrieved_cv_ids:
        top10_correct += 1

# Compute scores
top1_score = top1_correct / total_queries if total_queries > 0 else 0
top5_score = top5_correct / total_queries if total_queries > 0 else 0
top10_score = top10_correct / total_queries if total_queries > 0 else 0

# Print results
print("\nEvaluation Results:")
print(f"Total Queries: {total_queries}")
print(f"Top-1 Score: {top1_correct}/{total_queries} = {top1_score:.4f}")
print(f"Top-5 Score: {top5_correct}/{total_queries} = {top5_score:.4f}")
print(f"Top-10 Score: {top10_correct}/{total_queries} = {top10_score:.4f}")

# Save results to a file
results = {
    'Total Queries': total_queries,
    'Top-1 Correct': top1_correct,
    'Top-1 Score': top1_score,
    'Top-5 Correct': top5_correct,
    'Top-5 Score': top5_score,
    'Top-10 Correct': top10_correct,
    'Top-10 Score': top10_score
}
results_df = pd.DataFrame([results])
results_df.to_csv('rag_evaluation_results.csv', index=False)
print("\nResults saved to 'rag_evaluation_results.csv'")
```
